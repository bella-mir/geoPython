# V. Растровый анализ


## Темы

1. [**Растры как формат пространственных данных**](rasters_1.ipynb)

   Узнаете, как устроены растровые данные и научитесь работать с ними: читать, визуализировать, обрабатывать и подготавливать для анализа. На примере растра с портала WorldPop о распределнии населения в Тульской области

2. [**Зональная статистика**](rasters_2.ipynb)

   Научитесь агрегировать растровые данные по векторным зонам и рассчитывать показатели (например, плотность населения) для анализа территории.


## Задание: Оцениваем доступность для жителей

Каждый модуль заканчивается практическим заданием — этапом проекта **«15-минутный город»**, посвящённого анализу доступности повседневных функций. Проект можно выполнять для любого интересующего вас города.

В этом модуле мы блаблабла в исследуемом районе.

Задание оформлено в формате **Jupyter Notebook**, где предложена примерная последовательность шагов его выполнения. Оно доступно на платформе [Boosty](https://boosty.to/bellamir).

Там же размещены видео к каждому блоку и есть возможность задавать вопросы. Поддерживая проект, вы помогаете выпуску открытых материалов.
