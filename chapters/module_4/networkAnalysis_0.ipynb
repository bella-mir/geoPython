## Задание: Строим зоны доступности

Каждый модуль заканчивается практическим заданием — этапом проекта **«15-минутный город»**, посвящённого анализу доступности повседневных функций. Проект можно выполнять для любого интересующего вас города.

В этом модуле мы строим зоны **зоны пешеходной доступности 5, 10 и 15 минут** от выбранной категории объектов в исследуемом районе.
