# I. Пространственные данные


1. [**Модели пространственных данных**](spData_1.ipynb)

   Познакомитесь с двумя основными типами представления пространственных данных — векторным и растровым. Узнаете, в каком виде хранится информация и какие классы в Python используются для их описания.

2. **Чтение данных**

   Научитесь читать векторные данные из различных форматов и создавать набор пространственных данных на основе координат.

3. **Исследование набора данных**

   Исследуете один из наборов векторных данных: узнаете, как получить информацию о системе координат, количестве объектов, типе геометрии, полях и типах данных в них и др.

4. **Экспорт данных из OSM**

   Узнаете, как с помощью библиотеки `osmnx` экспортировать данные из OpenStreetMap по тегам и их значениям.
