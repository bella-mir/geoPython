# I. Пространственные данные


**Данные лежат в основе картографии и геопространственного анализа.**

В этом небольшом курсе мы начнём с изучения природы данных, с которыми работаем. Вы познакомитесь с моделями пространственных данных, научитесь загружать данные с локального компьютера, создавать пространственные объекты на основе координат, а также работать с OpenStreetMap (OSM) — одним из самых популярных открытых источников пространственных данных.

Этот модуль включает:

1. [**Модели пространственных данных**](module_11.ipynb)

   Вы узнаете о геометрических объектах — точках, линиях и полигонах.

2. [**Чтение данных**](module_12.ipynb)

   Вы научитесь загружать данные из различных форматов и типов файлов с использованием библиотек Python, таких как GeoPandas.

3. [**Исследование набора данных**](module_13.ipynb)

4. [**Экспорт данных из OSM**](module_14.ipynb)

   Вы узнаете, как получать и экспортировать данные OpenStreetMap .


## Чему вы научитесь:

- **Понимать структуру моделей пространственных данных**, знать, с помощью каких классов описываесявключая использование объектов Point, LineString и Polygon, и объяснять, как с их помощью представляются реальные географические объекты.

- **Работать с GeoDataFrame в библиотеке GeoPandas**, включая загрузку, просмотр и исследование наборов пространственных данных.

- **Читать пространственные данные** из различных форматов (например, Shapefile, GeoJSON, geoPackage), а также преобразовывать табличные данные с координатами в геопространственный формат.

- **Экспортировать пространственные данные из OpenStreetMap (OSM)** с помощью библиотеки OSMnx.
