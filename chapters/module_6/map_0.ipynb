# VI. Интерактивная визуализация


## Темы

1. [**Интерактивная карта**](map_1.ipynb)

   Научитесь создавать веб-карты с помощью Folium: добавлять несколько слоёв (землепользование, остановки, зоны доступности), настраивать стили, всплывающие подсказки и элементы управления.

2. [**Публикация на gh-pages**](map_2.ipynb)

   Узнаете, как сохранить карту в HTML и разместить её в интернете с помощью GitHub Pages, чтобы делиться результатами анализа по ссылке.


## Задание: Визуализируем результаты

Каждый модуль заканчивается практическим заданием — этапом проекта **«15-минутный город»**, посвящённого анализу доступности повседневных функций. Проект можно выполнять для любого интересующего вас города.

В этом модуле мы **создадим интерактивную веб-карту с результатами анализа**: отобразим основные слои, настроим их визуализацию и подготовим карту для публикации.

Задание оформлено в формате **Jupyter Notebook**, где предложена примерная последовательность шагов его выполнения. Оно доступно на платформе [Boosty](https://boosty.to/bellamir).

Там же размещены видео к каждому блоку и есть возможность задавать вопросы. Поддерживая проект, вы помогаете выпуску открытых материалов.
